```python
doc.owner_org.value_counts()
```

Ça me permet de savoir que la Défense Nationale est l'organisme gouvernemental faisant l'objet du plus grand nombre de demandes.

```python
defense = doc.owner_org == "dnd-mdn"
doc[defense].disposition.value_counts()
```

```python
doc[defense].disposition.value_counts().plot(kind="pie", title="Réponse du ministère de la Défense aux demandes d'accès")
```
